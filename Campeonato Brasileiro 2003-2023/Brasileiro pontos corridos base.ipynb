```python
import pandas as pd

base = pd.read_excel("Brasileiro pontos corridos.xlsx")

base_2023 = base.copy()

df_2023 = base_2023[base_2023['season'] == 2023]

df_2023_sorted = df_2023.sort_values(by=['points', 'won'], ascending=False)

base_2023.loc[base_2023['season'] == 2023] = df_2023_sorted.values

display(base_2023)

base_2023.to_excel("Brasileiro Pontos corridos 2003-2023.xlsx", index=False)
```
